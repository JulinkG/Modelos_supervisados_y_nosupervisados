## Variables

* ID: identificador de entrada
* Label: indicador si el cliente es considerado moroso o no
* Edad: edad del solicitante
* montoSolicitado: Monto que se solicita de crédito
* montoOtorgado: Monto que se le otorga de crédito
* Género: género de solicitante
* Quincenal:
* 1 si el crédito se solicita de pagos quincenales
* 0 si el crédito se solicita de pagos mensuales
* dependientesEconómicos: Número de dependientes económicos
* NivelEstudio: Nivel máximo de escolaridad
* score: score del solicitante.
* Ingresos Mensuales: ingresos mensuales de solicitante
* GastosMensuales: gastos mensuales del solicitante
* emailScore: Score de confiabilidad de su email
* Browser: Tipo de navegador que utiliza el solicitante
* NUMTDC_AV: número de tarjetas de crédito activas

## Selección y validación del modelo

```python
df_features.info()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_clas, Y_clas, test_size = 0.2, random_state= 7)
```

```python
model_LoR = LogisticRegression(solver="lbfgs",max_iter=1000)

model_LoR.fit(X_train,Y_train)
results = model_LoR.score(X_test,Y_test)
print(f'Accuracy Logistic Regresor: {results*100.0:,.2f}%')
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
models = []
models.append(('LoR', LogisticRegression(solver="lbfgs", max_iter= 1000)))
models.append(('K-NN', KNeighborsClassifier ()))
models.append(('SVC', SVC (gamma="auto")))
models.append(('NB', GaussianNB()))
```

```python
results = []
names = []
scoring = "accuracy"
```

```python
for name, model in models:
    kfold = KFold(n_splits=10)
    cv_results = cross_val_score(model,X_clas,Y_clas, cv = kfold, scoring = scoring )
    results.append(cv_results)
    names.append(name)

    print(f"{name}: {cv_results.mean()*100.0:,.2f}% ({cv_results.std()*100.0:,.2f}%)") 
```

```python
fig = plt.figure()
fig.suptitle("Comparacion de algoritmos")
ax=fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show
```
